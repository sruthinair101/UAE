# Create Categories

```python
# check for the number of unique ages in the age category
print("count of unique ages :" , cancer_df['Age'].nunique())
```

```python
print('\n')
# check for the unique ages in the age category
print('unique ages :',cancer_df['Age'].unique())
```

```python
print(cancer_df['Age'].describe())
```

```python
# create bins for the various age categories
min_age = cancer_df['Age'].min()
twenty_fifth = cancer_df['Age'].quantile(0.25)
median = cancer_df['Age'].median()
seventy_fifth = cancer_df['Age'].quantile(0.75)
maximum = cancer_df['Age'].max()
```

```python
# create a label and bins for the various age categories
labels = ["young_adult", "middle_age" , "older_adult", "elderlies"]
bins = [min_age, twenty_fifth, median, seventy_fifth, maximum]
```

```python
# create a category column for age
cancer_df['Age_cat'] = pd.cut(cancer_df['Age'], labels=labels, bins=bins)
```

```python
cancer_df[['Age','Age_cat']]
```

```python
column_list = ['Gender', 'Nationality', 'Emirate', 'Cancer_Type',
              'Outcome', 'Smoking_Status', 'Comorbidities', 'Ethnicity']

def plot_bar_graph():
  for c in column_list:
      g = sns.catplot(x=c, data=cancer_df, kind='count')
      g.set_xticklabels(rotation=90)

      #plt.title(c)
      plt.show()
plot_bar_graph()
```

## Exploring Relationships and Distributions

### Age Category Distribution

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Age_cat', data=cancer_df, palette='viridis', hue='Age_cat', legend=False)
plt.title('Distribution of Age Categories')
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.show()
```

### Outcome Distribution by Age Category

```python
plt.figure(figsize=(10, 7))
sns.countplot(x='Age_cat', hue='Outcome', data=cancer_df, palette='viridis')
plt.title('Outcome Distribution by Age Category')
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Outcome')
plt.show()
```

### Outcome Distribution by Gender (Pie Charts)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



gender_outcomes = cancer_df.groupby('Gender')['Outcome'].value_counts(normalize=True).unstack()

fig, axes = plt.subplots(1, len(gender_outcomes.index), figsize=(18, 6))

for i, gender in enumerate(gender_outcomes.index):
    ax = axes[i]
    gender_outcomes.loc[gender].plot.pie(autopct='%1.1f%%', startangle=90, ax=ax, colors=sns.color_palette('viridis', len(gender_outcomes.columns)))
    ax.set_title(f'Outcome Distribution for {gender}')
    ax.set_ylabel('') # Hide the default 'Outcome' label

plt.tight_layout()
plt.show()
```

## Correlation Matrix for Numerical Features

```python
numerical_cols = ['Age', 'Weight', 'Height']
correlation_matrix = cancer_df[numerical_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
